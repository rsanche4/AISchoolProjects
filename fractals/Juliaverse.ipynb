```python
# Second attempt
# A beautiful animation of JuliaSet fractals as we change c
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
import tqdm
import cv2
import os

def juliaset(z0, c, max_iter):
    z = z0
    n = 0
    while abs(z) <= 2 and n < max_iter:
        z = z**2 + c
        n += 1
    return n  # Only return the iteration count

def compute_julia_set(xvals, yvals, c, max_iter=50):
    # Create empty 2D array for the image
    img = np.zeros((len(yvals), len(xvals)))
    
    for i, y in enumerate(yvals):
        for j, x in enumerate(xvals):
            img[i, j] = juliaset(x + 1j * y, c, max_iter) / max_iter  # Normalized intensity
    
    return img


# Convert your matplotlib images to OpenCV format
def prepare_frame(img, cmap='magma'):
    """Convert a matplotlib-style image to BGR for OpenCV"""
    # Normalize to 0-255
    normalized = (img - img.min()) / (img.max() - img.min()) * 255
    normalized = normalized.astype(np.uint8)
    
    # Apply colormap
    if cmap == 'magma':
        colored = cv2.applyColorMap(normalized, cv2.COLORMAP_MAGMA)
    else:
        colored = cv2.cvtColor(normalized, cv2.COLOR_GRAY2BGR)
    return colored

# Parameters
xvals = np.linspace(-2.0, 2.0, 640)
yvals = np.linspace(-2.0, 2.0, 640)
max_iter = 50

i = np.linspace(-5.0, 5.0, 250)
j = np.linspace(-5.0, 5.0, 250)
#c = complex(0.25, 0.7) 

frame_count = 0
save_dir = "JuliaImgs"
for ci in tqdm.tqdm(i):
    for reversal in range(2):
        if reversal==0:
            for cj in j:
                c = complex(ci, cj)
                julia_img = compute_julia_set(xvals, yvals, c, max_iter)
                frame = prepare_frame(julia_img)
                
                filename = os.path.join(save_dir, f"{frame_count:06d}.png")
                cv2.imwrite(filename, frame)
                frame_count += 1
        else:
            for cj in j[::-1]: # go back to the beginning
                c = complex(ci, cj)
                julia_img = compute_julia_set(xvals, yvals, c, max_iter)
                frame = prepare_frame(julia_img)

                filename = os.path.join(save_dir, f"{frame_count:06d}.png")
                cv2.imwrite(filename, frame)
                frame_count += 1
```
