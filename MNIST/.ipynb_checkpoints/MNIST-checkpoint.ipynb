```python
import numpy as np
import gzip
import struct
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets


def read_idx(filename):
    with gzip.open(filename, 'rb') as f:
        zero, data_type, dims = struct.unpack('>HBB', f.read(4))
        shape = tuple(struct.unpack('>I', f.read(4))[0] for d in range(dims))
        return np.frombuffer(f.read(), dtype=np.uint8).reshape(shape)
```

```python
# here I am readinig the images

X_train_filename = "MNIST/train-images-idx3-ubyte.gz"
y_train_filename = "MNIST/train-labels-idx1-ubyte.gz"
X_testing_filename = "MNIST/t10k-images-idx3-ubyte.gz"
y_testing_filename = "MNIST/t10k-labels-idx1-ubyte.gz"

X_train = read_idx(X_train_filename)
y_train = read_idx(y_train_filename)
X_test = read_idx(X_testing_filename)
y_test = read_idx(y_testing_filename)
```

```python
# ok that is correct. They match to the correct digit, thus, lets move on.
# So for the next we need to actually create our activation and neurons and network. We are going to have 28*28 = 784 for the inputs so a flattened pixel values. Then we need two hidden layers, each let's have them be 16 neurons. Using 
# sigmoid and
# and finally 10 output neurons using softmax.
# (784, 16) (16, 16) (16, 10)

# let us define our sigmoid, and our softmax
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

print(sigmoid(2))
```

```python
import numpy as np

def softmax(x):
    e_x = np.exp(x - np.max(x))  # subtract max for numerical stability
    return e_x / e_x.sum()

logits = np.array([2.0, 1.0, 0.1]) # an example
probabilities = softmax(logits)
print(probabilities)  # Output: array([0.659, 0.242, 0.099])
```

```python
import numpy as np

np.random.seed(695)

class DigitClassifier:
    def __init__(self):
        # Initialize weights and biases
        self.weights1 = np.random.randn(784, 128) * np.sqrt(2./784)  # Xavier/Glorot initialization
        self.bias1 = np.zeros(128)
        self.weights2 = np.random.randn(128, 64) * np.sqrt(2./128)
        self.bias2 = np.zeros(64)
        self.weights3 = np.random.randn(64, 10) * np.sqrt(2./64)
        self.bias3 = np.zeros(10)
        
        # Cache for backpropagation
        self.cache = {}
    
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    
    def sigmoid_derivative(self, z):
        s = self.sigmoid(z)
        return s * (1 - s)
    
    def softmax(self, z):
        exp_z = np.exp(z - np.max(z))
        return exp_z / np.sum(exp_z)
    
    def categorical_crossentropy(self, y_true, y_pred):
        n_samples = y_true.shape[0]
        y_pred_clipped = np.clip(y_pred, 1e-12, 1 - 1e-12)
        return -np.sum(y_true * np.log(y_pred_clipped)) / n_samples
    
    def feed_forward(self, X):
        # Layer 1
        z1 = np.dot(X, self.weights1) + self.bias1
        a1 = self.sigmoid(z1)
        
        # Layer 2
        z2 = np.dot(a1, self.weights2) + self.bias2
        a2 = self.sigmoid(z2)
        
        # Output layer (logits)
        z3 = np.dot(a2, self.weights3) + self.bias3
        a3 = self.softmax(z3)
        
        # Store values for backpropagation
        self.cache = {'X': X, 'z1': z1, 'a1': a1, 'z2': z2, 'a2': a2, 'z3': z3, 'a3': a3}
        
        return a3
    
    def backward_pass(self, y_true_i, learning_rate=0.01):
        # Retrieve cached values (ensure 2D for matrix ops)
        X = self.cache['X'].reshape(1, -1)  # (1, 784)
        z1 = self.cache['z1'].reshape(1, -1)  # (1, 128)
        a1 = self.cache['a1'].reshape(1, -1)  # (1, 128)
        z2 = self.cache['z2'].reshape(1, -1)  # (1, 64)
        a2 = self.cache['a2'].reshape(1, -1)  # (1, 64)
        a3 = self.cache['a3'].reshape(1, -1)  # (1, 10)
        y_true_i = y_true_i.reshape(1, -1)  # (1, 10)
    
        # Output gradient
        dz3 = a3 - y_true_i  # (1, 10)
    
        # Layer 3 Gradient
        dw3 = np.dot(a2.T, dz3)  # (64, 10)
        db3 = np.sum(dz3, axis=0)  # (10,)
    
        # Layer 2 gradient
        dz2 = np.dot(dz3, self.weights3.T) * self.sigmoid_derivative(z2)  # (1, 64)
        dw2 = np.dot(a1.T, dz2)  # (128, 64)
        db2 = np.sum(dz2, axis=0)  # (64,)
    
        # Layer 1 gradient
        dz1 = np.dot(dz2, self.weights2.T) * self.sigmoid_derivative(z1)  # (1, 128)
        dw1 = np.dot(X.T, dz1)  # (784, 128)
        db1 = np.sum(dz1, axis=0)  # (128,)
    
        # Update parameters
        self.weights3 -= learning_rate * dw3
        self.bias3 -= learning_rate * db3
        self.weights2 -= learning_rate * dw2
        self.bias2 -= learning_rate * db2
        self.weights1 -= learning_rate * dw1
        self.bias1 -= learning_rate * db1
    
    def train(self, X_flattened, y_one_hot, batch_size=128, epochs=100, learning_rate=0.01):
        num_samples = X_flattened.shape[0]
        indices = np.arange(num_samples)
        loss_hist = []
        
        for epoch in range(epochs):
            np.random.shuffle(indices)
            X_shuffled = X_flattened[indices]
            y_shuffled = y_one_hot[indices]
            
            X_batch = X_shuffled[:batch_size]
            y_batch = y_shuffled[:batch_size]
            
            # Forward pass
            y_pred = np.zeros((batch_size, 10))
            for i, image in enumerate(X_batch):
                #print(image.shape)
                #print(y_batch[i])
                preds_logits = self.feed_forward(image)
                y_pred[i] = preds_logits

                # Backward pass
                self.backward_pass(y_batch[i], learning_rate)
            
            # Compute loss
            loss = self.categorical_crossentropy(y_batch, y_pred)
            loss_hist.append(loss)
            
            print(f"Epoch: {epoch+1}/{epochs}, Loss: {loss:.4f}")
        
        return loss_hist

```

```python
classifierDigits = DigitClassifier()
to_plot_loss_hist = classifierDigits.train(X_train, y_train_one_hot)
```

```python
plt.plot(to_plot_loss_hist)
plt.title("Loss History")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
X_test = X_test.reshape(10000, 28 * 28)
y_test = np.eye(num_classes)[y_test]
```

```python
# now let us perform some checks with the test datasets see how well it's performing!
accuracy = 0
all_tests = y_test.shape[0]
ytest = []
preds = []
showdisplay_count = 0 # lets only show 10 failed digits as examples
for ind in range(all_tests): # There is about 10000 diffeernt tests
    
    index = ind
    preds_logits = classifierDigits.feed_forward(X_test[index])
    true_digit = np.argmax(y_test[index])
    pred_digit = np.argmax(preds_logits)
    if true_digit==pred_digit:
        accuracy+=1
    elif showdisplay_count < 10:
        # lets display the failed digits!
        image_to_show = X_test[index].reshape(28, 28)
        plt.figure(1, figsize=(3, 3))
        plt.title(f"True Label: {true_digit} | Pred Label: {pred_digit}")
        plt.imshow(image_to_show, cmap="gray", interpolation="nearest")
        plt.show()
        showdisplay_count+=1
    
    # just to see which of digits are weird, good for visualizing confusion matrix
    ytest.append(true_digit)
    preds.append(pred_digit)

print(f"Accuracy: {int(np.ceil((accuracy/all_tests)*100))}%")
```

```python
# That's good enough for me!
# Now lets see which ones are getting misclassified more
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(ytest, preds)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", 
            xticklabels=range(10), yticklabels=range(10))
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```
